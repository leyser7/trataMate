```python
# Takes in a dataframe, finds the most correlated variables with the
# grade and returns training and testing datasets
def format_data(df):
    # Targets are final grade of student
    labels = df['Fertility']
    
    # Drop the school and the grades from features
    #df = df.drop(columns=['school', 'G1', 'G2', 'percentile'])
    
    # One-Hot Encoding of Categorical Variables
    #df = pd.get_dummies(df)
    
    # Find correlations with the Grade
    most_correlated = df.corr().abs()['Fertility'].sort_values(ascending=False)
    
    # Maintain the top 6 most correlation features with Grade
    most_correlated = most_correlated[:8]
    
    df = df.ix[:, most_correlated.index]
    #df = df.drop(columns = 'Premium_yes')
    #df = df.drop(columns = 'Premium_no')
    
    # Split into training/testing sets with 25% split
    X_train, X_test, y_train, y_test = train_test_split(df, labels, 
                                                        test_size = 0.25,
                                                        random_state=42)
    return X_train, X_test, y_train, y_test
```

```python
# Splitting data into training/testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = format_data(df)
X_train.head()
```

```python
# Create the features and response
X = df.loc[:, ['Catholic','Infant.Mortality']]
y = df.loc[:, 'Fertility']
```

# Minimos cuadrados:

\begin{equation}\hat{\boldsymbol{\beta}}=\left(\mathbf{X}^{\top} \mathbf{X}\right)^{-1} \mathbf{X}^{\top} \mathbf{y}\end{equation}

```python
import numpy as np
def linear_regression(X,y):
    inv = np.linalg.inv(X.T.dot(X))
    temporal = inv.dot(X.T)
    _coeffs = temporal.dot(y)
    return _coeffs
```

```python
# Run the by hand implementation
by_hand_coefs = linear_regression(X, y)
print('Intercept calculated by hand:', by_hand_coefs[0])
print('Slope calculated by hand: ', by_hand_coefs[1])
```

```python
xs = np.linspace(4, 31, 4000)
print(xs.shape)
ys = by_hand_coefs[0] + by_hand_coefs[1] * xs

plt.figure(figsize=(8, 8))

plt.plot(df['Infant.Mortality'], df['Fertility'], 'bo',label = 'observations', alpha = 0.8);
plt.xlabel('Infant.Mortality', size = 18); plt.ylabel('Price', size = 18);
plt.plot(xs, ys, 'r--', label = 'OLS Fit', linewidth = 3)
plt.legend(prop={'size': 16})
plt.title('Calories burned vs Duration of Exercise', size = 20);
```

# Regresion Bayesiana Lineal

```python
with pm.Model() as linear_model:
    # Intercept
    intercept = pm.Normal('Intercept', mu = 0, sd = 10)
    
    # Slope 
    slope = pm.Normal('slope', mu = 0, sd = 10)
    
    # Standard deviation
    sigma = pm.HalfNormal('sigma', sd = 10)
    
    # Estimate of mean
    mean = intercept + slope * X.loc[:, 'Infant.Mortality']
    
    # Observed values
    Y_obs = pm.Normal('Y_obs', mu = mean, sd = sigma, observed = y.values)
    
    # Sampler
    step = pm.NUTS()

    # Posterior distribution
    linear_trace = pm.sample(1000, step)
```

```python
pm.traceplot(linear_trace, figsize = (12, 12));
```

```python
pm.plot_posterior(linear_trace, figsize = (12, 10));
```

```python
import numpy as np
plt.figure(figsize = (8, 8))
pm.plot_posterior_predictive_glm(linear_trace, samples = 100, eval=np.linspace(2, 30, 100), linewidth = 1, 
                                 color = 'red', alpha = 0.8, label = 'Bayesian Posterior Fits',
                                lm = lambda x, sample: sample['Intercept'] + sample['slope'] * x);
plt.scatter(X['Infant.Mortality'], y.values, s = 12, alpha = 0.8, c = 'blue', label = 'Observations')
plt.plot(X['Infant.Mortality'], by_hand_coefs[0] + X['Infant.Mortality'] * by_hand_coefs[1], 'k--', label = 'OLS Fit', linewidth = 1.4)
plt.title('Posterior Predictions with all Observations', size = 20); plt.xlabel('Duration (min)', size = 18);
plt.ylabel('Calories', size = 18);
plt.legend(prop={'size': 16});
```

```python
import seaborn as sns
value = 30.0
bayes_prediction = linear_trace['Intercept'] + linear_trace['slope'] * value
plt.figure(figsize = (8, 8))
plt.style.use('fivethirtyeight')
sns.kdeplot(bayes_prediction, label = 'Bayes Posterior Prediction')
plt.vlines(x = by_hand_coefs[0] + by_hand_coefs[1] * value, 
           ymin = 0, ymax = 0.085, 
           label = 'OLS Prediction',
          colors = 'red', linestyles='--')
plt.legend();
plt.xlabel('Fertility', size = 18), plt.ylabel('Probability Density', size = 18);
plt.title('Posterior Fertility for 30 Infant.Mortality', size = 20);
```
